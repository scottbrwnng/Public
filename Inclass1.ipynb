```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df= pd.read_csv('/Users/scott/Downloads/countries.csv')

df.columns
```

```python
#smallest 10 populations
smallest_pop = df[['country', 'population']].nsmallest(10, 'population')
smallest_pop
```

```python
#largest 10 populations
largest_pop = df[['country', 'population']].nlargest(10, 'population')
largest_pop
```

```python
country_group = df.groupby(['country'])
country_group.get_group('China')
```

```python
country_group.get_group('United States')
```

```python
ireland_df = df[df['country'] == 'Ireland']
ireland_df
```

```python
mexico_df = df[df['country'] == 'Mexico']
mexico_df
```

```python
us_df = df[df['country'] == 'United States']
us_df
```

```python
xs = []
ys = []
mex_ys = []
us_ys = []
for x in ireland_df['year']:
    xs.append(x)
for y in ireland_df['population']:
    ys.append(y)
for i in mexico_df['population']:
    mex_ys.append(i)
for u in us_df['population']:
    us_ys.append(u)
plt.plot(xs, ys)
plt.plot(xs, mex_ys)
plt.plot(xs, us_ys)
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend(["Ireland", "Mexico", "United States"])
plt.show()
```
