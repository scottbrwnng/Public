```python
import matplotlib.pyplot as plt

xs = [1,2,3,4,5]
ys = []
negys = []

for x in xs:
    ys.append(x**2)
    negys.append(-x**2)
plt.plot(xs, ys)
plt.plot(xs, negys)
plt.xlabel("The x axis")
plt.ylabel("The y axis")
plt.axhline(0)
plt.axvline(3)
plt.title("This is an example graph")
plt.legend(["positive ys", "negative ys"])
fig = plt.gcf()
fig.savefig('/users/scott/desktop/newfig.png')
plt.show()
```
