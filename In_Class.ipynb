```python
import pandas as pd 
import numpy as np
filename = '/Users/scott/Downloads/survey_results_public.csv'
pd.options.display.float_format = '${:,.0f}'.format
df = pd.read_csv(filename, index_col = 'ResponseId')
df.columns.value_counts()

```

```python
#Lets do one together

#Group the results by country
country_group = df.groupby(['Country'])
```

```python
#Review the result for Ireland
country_group.get_group('Ireland')
```

```python
#Calculate Selected statistics for the entire dataframe
country_group.ConvertedCompYearly.agg(['mean', 'mean', 'std', 'max', 'min'])
```

```python
#Selected statistics for the countries we are interested in
int_list = ['India', 'Ireland', 'United States of America']
country_group.ConvertedCompYearly.agg(['mean', 'median', 'std', 
'max', 'min']).loc[int_list]
```

```python
#Review the result for Ireland
country_group.get_group('Ireland')
```

```python
# Sort the median compensation totals by country
country_agg = country_group.ConvertedCompYearly.agg(['median', 
'mean', 'std', 'max', 'min'])
country_agg.sort_values('median', ascending=False)
country_agg.nlargest(10, 'median')
country_agg.nsmallest(10, 'median')
                     
```
